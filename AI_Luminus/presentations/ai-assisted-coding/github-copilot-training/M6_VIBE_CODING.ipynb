---
## What is vibe coding?

Vibe coding means driving an AI agent with intent, not instructions. Instead of writing a precise spec, you describe what you want to see and iterate from the output. The agent writes the code; you review and steer.

This is a different posture from M2–M5 where every prompt had precise context and acceptance criteria. Here the goal is deliberately open — and observing how the agent handles ambiguity is part of the exercise.
