## Key teaching moment

**Copilot can explain a function correctly AND miss what it silently passes through.**

Deliver this at Q2, after participants have run the code and seen identical results across seasons:

> "`filter_valid_curves` rejects bad types, bad quotes, unknown units — ask Copilot and it tells you correctly. Then ask: what does it pass through without checking? It will hesitate. The `cetStarttime` column is empty in the input. The loader doesn't touch it, validate it, or warn about it. That gap is exactly where the bug lives."

Secondary teaching moment at `/init` (which comes *before* Q1):

> "Before you read a single function — ask Copilot to describe the whole repo. What `/init` generates tells you what AI understands at a glance, and what it misses. Does it mention `cetStarttime`? Does it describe what happens if that column is empty? That gap between what Copilot knows and what the bug requires is what we're building toward all day."

The `/doc` closing moment — after Q2:

> "You just discovered that `filter_valid_curves` passes through records with empty CET columns. Now generate a docstring for it. Does the docstring mention that gap? Almost certainly not. The docstring describes what the function *does*. It doesn't describe what it *doesn't do*. That's the gap tests are for."

## Questions and answers

**Q1: What do `load_curves_csv` and `filter_valid_curves` do?**

A: `load_curves_csv` reads the CSV and returns raw string dicts — no type conversion. It raises if the file is missing or required columns are absent.

`filter_valid_curves`:

Drops records where `type` is not in {CALCULATED, INPUT}, `quote` is non-numeric, or `unit` is unrecognised. It passes through everything else silently — including records with missing or empty CET columns.

---

**Q2: What does `filter_valid_curves` silently pass through?**

A: Records with empty `cetStartDate` and `cetStarttime`. The loader has no awareness of these columns — it loads and filters on structural validity only. The copy-through bug produces values in those columns (not empty, just wrong), so it would also pass this filter.

---

**After `/init`:**

Q: "What did `/init` generate that surprised you?"

A: Open answer. Common: it described the regression runner but didn't mention the CET columns; or it described the file structure but not the data flow. Either is a good discussion about what AI infers vs. what must be stated explicitly.

---

**After `/doc`:**

Q: "What's missing from the docstring that you just discovered in Q2?"

A: That `filter_valid_curves` passes through records with missing CET columns. The docstring describes what it rejects; it won't mention the gap unless the function explicitly handles it.

## Watch for

**Participants who open `cet_converter.py` in M1.**

What it looks like: They ask Copilot about the CET conversion or open the file directly.

Response: Close it. "M1 is loader only. You'll see `cet_converter.py` for the first time in M2 — when the regression tells you something is wrong."

---

**Participants who find the bug in M1 via `/init` or `/doc`.**

What it looks like: `/init` generates a CLAUDE.md that mentions the copy-through bug, or Copilot surfaces it unprompted.

Response: Acknowledge and park it. "Good catch. Hold that. We're going to feel the pain of not finding it first, then fix it properly."

---

**`/init` generates a very long CLAUDE.md.**

What it looks like: A 300-line file covering every directory.

Response: Ask: "Is this useful to a new engineer? What would you cut?" The trimming exercise surfaces what the participant thinks is essential context.
