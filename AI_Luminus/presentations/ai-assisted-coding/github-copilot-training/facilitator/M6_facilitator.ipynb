# M6 — Vibe Coding: Curve Dashboard
**Duration:** 45 min | **Participant notebook:** `M6_VIBE_CODING.ipynb`

---

## At a glance

| | |
|---|---|
| **Copilot skill taught** | Open-ended agent prompting, intent vs spec, iterating toward a target |
| **Key files** | `dashboard_mockup.html` (target), `regression/actual/` (generated), `dashboard.html` (produced) |
| **What participants produce** | `dashboard.html` — a self-contained browser dashboard showing actual vs checkpoint mismatches |

**The mockup is the acceptance criteria.** `dashboard_mockup.html` shows the target: summary banner (0/75/4), line charts with the 1h gap visible, mismatch rows in red, order B called out. Participants iterate their agent prompts until they get close to it.

**The lesson is the gap** between the agent's first attempt and the mockup — not the final dashboard quality.

## Key teaching moment

**The gap between what the agent produces and the mockup is the lesson.**

Deliver this before Step 2 (first agent run):

> "You have seen the target. Now give the agent one short sentence and see what it builds. Do not describe the mockup. Just describe the goal. The difference between what it produces and what you want — that is what you will close through iteration."

At Step 3 (review):

> "Before you ask it to improve — is the data correct? Did it read the actual CSV files, or did it invent plausible-looking values? An agent that ships confidently from wrong data is more dangerous than one that says it doesn't know."

At Step 4 (iterate):

> "In M4, every prompt had a file reference and a test. Here the prompt is one sentence. What did the agent fill in that you didn't specify? Was that a good guess or a bad one?"

## Questions and answers

Q: "In M4 you gave the agent precise tests and a spec. Here you gave it intent. Which produced better output?"

A: No single right answer — this is intentionally comparative. Precise spec → more predictable output, easier to verify, harder to write. Intent → faster to start, agent makes choices you may not want, harder to verify correctness. The lesson: match the approach to what you are building. Novel UI → vibe coding is fine. Business logic with correctness requirements → TDD.

---

Q: "Did the agent produce something visually convincing but technically wrong?"

A: Often yes — agents will generate plausible-looking HTML with hardcoded or averaged values rather than actual CSV data, especially if the file-reading step is complex. This is the most important catch in M6. If participants didn't check the data, ask: "How do you know the values in the dashboard match the CSVs?"

---

Q: "When is vibe coding the right approach?"

A: When the output is easy to verify visually (UI, charts), when correctness is best-effort rather than exact, when speed matters more than precision. Wrong for: financial calculations, CET conversions, anything with a regression test. Right for: dashboards, prototypes, tooling, documentation.

## Watch for

**Agent builds a dashboard with hardcoded or invented data.**

What it looks like: The HTML is visually polished but the values don't match the CSVs.

Response: This is a teaching moment, not a failure. "How would you verify this? What's the simplest check?" Answer: run the cell that prints CSV rows and compare manually, or ask the agent to add a data source indicator.

---

**Agent generates a React/Vue/Webpack setup instead of plain HTML.**

What it looks like: It creates `package.json`, `webpack.config.js`, and a build step.

Response: The prompt said browser-native, no build step. Ask: "Did it follow the constraint? How would you re-prompt to prevent this?" Then show: adding "use only vanilla HTML, CSS, and inline JavaScript — no npm, no build step" makes the constraint explicit.

---

**Participants' first attempt is already close to the mockup.**

What it looks like: The dashboard matches the mockup after one prompt.

Response: Push to Step 5 (stress test prompts). The ambiguous prompt ("make it more insightful") almost always produces something interesting to discuss.

---

**Participants spend all their time iterating and never reach the debrief.**

What it looks like: They are still refining the dashboard at 14:15.

Response: Cut them off at 14:10 regardless. The debrief questions are more valuable than another iteration. Say: "Whatever you have is fine — the dashboard is not the point. The questions at the end are."

**Closing the day (30 min after M6):**

Ask around the room: "We started with 29,211 mismatches invisible for 10 months. Walk me back through the modules — what would have caught it?"

Expected answers: M1 (understood the implicit assumptions), M2 (precise context catches what vague prompts miss), M3 (encoded the rules permanently), M4 (summer and fall-back tests), M5 (SODA catches shape problems), M6 (knowing when to verify agent output vs trust it).

Closing line:
> "Copilot is fast. It is also confidently wrong in exactly the cases where being right matters most. The regression runner, the tests, the SODA checks — those are not bureaucracy. They are the thing that makes 'it works on my machine' mean 'it works in June in production'."
