## Key teaching moment

**Custom instructions turn tribal knowledge into persistent context. Specificity is what makes them useful.**

### Opening concept — instructions vs prompts (5 min)

Walk through this distinction before participants start:

| | `copilot-instructions.md` | `.github/prompts/` file |
|---|---|---|
| When it applies | Every prompt automatically | Only when invoked with `/command` |
| Good for | Project-wide rules, domain vocabulary | Specific repeatable tasks |
| Written by | The team, once | Anyone, task by task |

> "In M2 you had to repeat the CET rule in every prompt. After M3, it lives in `copilot-instructions.md` — you never need to type it again. But not everything belongs there. A one-off task — review this function, scaffold this wrapper — belongs in a prompt file you invoke when you need it."

---

### At Step 1 — why specificity matters

> "Everything we just learned in M2 — the three CET regimes, order B, negative quotes valid — lives in your chat history right now. Tomorrow it's gone. A new engineer joins: gone for them too. Custom instructions are how you make Copilot as domain-aware as your most experienced engineer, every session."

Deliver the specificity point live. Show two versions:
- Vague: *"Be careful with timezones"*
- Precise: *"In the October DST transition, records with order=B at 02:00 are already in CET — do not apply the CEST offset"*

Ask Copilot to write a CET function with each version active. The difference makes the point.

### At Step 3 — instructions vs prompts again

> "An instruction tells Copilot what to *know*. A prompt tells it what to *do*. The instructions are always active; the prompt is invoked when you need a specific task. You've already seen `/review-cet-function` — that's the same pattern. Now you write one for a task that matters to your work."

## Questions and answers

**After Step 2 (test the instructions):**

Q: "Did Copilot mention the three CET regimes without you prompting it?"

A: If the instructions are specific enough, yes. If it still doesn't mention them, the instruction is too vague. Ask: "Show me the exact text in your instructions file about the regimes. Is it a table, or a sentence?"

---

Q: "If these instructions had been in place before the original SAS migration, which mismatches would they have prevented?"

A: All of the summer ones — the CEST offset rule would have been in context when the migration code was written. Order B is less clear: the rule needs to be precise enough that Copilot applies it correctly. Good discussion, not a clean answer.

---

Q: "What belongs in `copilot-instructions.md` versus a `.github/prompts/` file?"

A: Instructions = what Copilot should *know* always. Prompts = a specific task you invoke. Rule of thumb: if you'd want every prompt to respect it, it goes in instructions. If you'd invoke it for a specific task (review a function, scaffold a wrapper), it's a prompt.

---

**After `/create-prompt`:**

Q: "Does your new prompt reference `#file:` paths? Should it?"

A: A project-specific prompt should — it makes the task precise. A skill (`.github/skills/`) shouldn't, because it's designed to work in any repo. This distinction is the setup for M5.

## Watch for

**Participants who write vague instructions.**

What it looks like: "Handle timezones carefully." "Be aware of CET/CEST."

Response: Run the live demo. Ask Copilot to write a CET conversion function with those instructions. It will produce something plausible but wrong. Then add the specific three-regime table and run the same prompt. The difference is visible immediately.

---

**Participants who write instructions that are too long.**

What it looks like: A 200-line instructions file with section headers, full code examples, and commentary.

Response: "Copilot reads this on every prompt. How many of these lines are actually constraints vs. documentation?" Push them to trim to rules only — 10–20 lines is plenty.

---

**`/create-prompt` generates a prompt that duplicates instructions content.**

What it looks like: The prompt file repeats all the CET rules that are already in `copilot-instructions.md`.

Response: Ask: "Does a prompt file need to repeat what's already in the instructions? What happens if you remove the duplicated rules and rely on instructions for context?" The answer: prompts should be task-focused, not context-duplicating. The instructions handle always-on context.

## Timing notes

| Segment | Planned | Actual typical | Notes |
|---|---|---|---|
| Intro — instructions vs prompt table | 5 min | 5 min | Keep short; the distinction lands better after they've built both |
| Step 1 — generate instructions | 8 min | 8 min | Draft is usually decent; refinement takes longer |
| Step 2 — test instructions | 7 min | 10 min | Participants iterate when the first test doesn't show the regimes |
| Step 3 — `/create-prompt` | 10 min | 10 min | Choosing what task to automate takes longer than generating the prompt |

**M3 is the shortest module and it's post-break.** Keep energy up. The live vague-vs-precise demo is the most engaging moment — do it at the front if you have time.

**If you're running long:** Skip Step 3 (`/create-prompt`). Step 1 + Step 2 (write and test instructions) are the core. `/create-prompt` is valuable but skippable; participants will use it in M4 anyway when they write their own review prompt.

**If you're running short:** Ask: "What would you add to the instructions that would prevent the order B mistake Copilot made in M2 Step 2?" Writing that one rule together takes 5 minutes and is a strong close.
