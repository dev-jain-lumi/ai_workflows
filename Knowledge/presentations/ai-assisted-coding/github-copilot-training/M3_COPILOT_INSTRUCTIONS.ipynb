## What is `copilot-instructions.md`?

The file `.github/copilot-instructions.md` is read automatically by VS Code Copilot at the start of every chat. It is **always-on context** — you never need to add `#file:` references to it.

This is different from an **agent prompt** (a `.github/prompts/` file), which you invoke explicitly with a slash command when you need it for a specific task.

| | `copilot-instructions.md` | `.github/prompts/` file |
|---|---|---|
| When it applies | Every prompt in this repo | Only when you invoke it with `/command` |
| Good for | Project-wide rules, domain vocabulary, non-negotiable constraints | Specific repeatable tasks (review, scaffold, generate) |
| Written by | The team, once | Anyone, task by task |

In M2 you had to repeat the CET rule in every prompt. After M3, it lives in `copilot-instructions.md` — you never need to type it again.

## Step 3 — Create a reusable prompt with `/create-prompt`

You have written instructions that apply to the whole project. Now write a **prompt** for a specific, repeatable task.

A `.github/prompts/` file is invoked with a slash command whenever you need it — not on every prompt. It can be shared with your team via git.

**Think of a task you would repeat on every new curve migration** — for example:
- Reviewing a CET conversion function
- Generating a regression checkpoint coverage check
- Scaffolding a new DC wrapper from the existing pattern

Describe the task you want to automate. Copilot will generate a `.github/prompts/<name>.prompt.md` file.

**Ask Copilot Chat:**



```python
'''
/create-prompt implement a pipeline stage function following project conventions, then verify with regression.
'''
```

Review the generated file. Does it reference the right files? 

✅ Your new slash command appears in Copilot Chat autocomplete. Try it out :

```python
'''
/<your_prompt_name> add a filter_by_quoted_date function to curve_loader
'''
```
